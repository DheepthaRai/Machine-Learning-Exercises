```python
import sklearn
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/servo/servo.data"
df = pd.read_csv(url, names=['motor', 'screw', 'p_gain', 'v_gain', 'Class'])
df.head()
```

```python
df.motor.replace({'A':0,'B':1,'C':2,'D':3,'E':4},inplace=True)
df.screw.replace({'A':0,'B':1,'C':2,'D':3,'E':4},inplace=True)
```

```python
df.head()
```

```python
X = df.drop('Class',axis=1)
y = df.Class
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 3)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)
```

```python
y_pred = regressor.predict(X_test)
y_pred_train = regressor.predict(X_train)
# printing model score
print("Model Score: ", regressor.score(X_test, y_test))
# printing r_square_score
from sklearn.metrics import r2_score
print("R_square score: ", r2_score(y_test,y_pred))
```

```python
# Visualising the Test set results
plt.scatter(X_train.motor, y_train, color = 'red')
plt.plot(X_train.motor, regressor.predict(X_train), color = 'blue')
plt.title('Motor vs Class (Training set)')
plt.xlabel('Motor Type')
plt.ylabel('Class')
plt.show()
```

```python
plt.scatter(X_train.screw, y_train, color = 'red')
plt.plot(X_train.screw, regressor.predict(X_train), color = 'blue')
plt.title('Screw vs Class (Training set)')
plt.xlabel('Screw Type')
plt.ylabel('Class')
plt.show()
```

```python
plt.scatter(X_train.p_gain, y_train, color = 'red')
plt.plot(X_train.p_gain, regressor.predict(X_train), color = 'blue')
plt.title('P_Gain vs Class (Training set)')
plt.xlabel('P_Gain Type')
plt.ylabel('Class')
plt.show()
```
